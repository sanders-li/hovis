```python
import json
import pandas as pd
```

```python
filename='reported_hospital_capacity_admissions_facility_level_weekly_average_timeseries_20210103_geocoded.json'
df = pd.read_json(filename, orient='records')
```

```python
df
```

```python
res = df.groupby('collection_week').apply(lambda row: json.loads(row.to_json(orient='records', indent=4)))
res
```

```python
out = {}
for k,v in res.iteritems():
    out[k] = v
```

```python
filename = 'reported_hospital_capacity_admissions_facility_level_weekly_average_timeseries_20201215_metrics_grouped.json'
with open(filename, 'w+') as f:
    f.write(json.dumps(out, indent=4))
print('Done')
```
